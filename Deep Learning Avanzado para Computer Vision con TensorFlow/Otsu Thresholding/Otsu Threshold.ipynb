# Otsu Method for Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('penguin.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
otsu_Threshold,_ = cv2.threshold(img, 0, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
bw_img = np.uint8(255*(img < otsu_Threshold))

fig = plt.imshow(bw_img, cmap = 'gray')
plt.grid(False)
plt.axis('off')
plt.show()
```
